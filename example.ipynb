```python
import pandas as pd
from aijobstrends import AITrendsAnalyzer 
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt

print("Initializing AIJobsTrends Analyzer...")

data_path = 'aijobstrends/data/sample_data.csv'

try:
    analyzer = AITrendsAnalyzer(file_path=data_path)
    print("Analyzer initialized and data loaded successfully.")
    print("-" * 40)
except Exception as e:
    print(f"FATAL ERROR: Initialization failed. Please check the CSV file and column names: {e}")
```

```python
print("1. Calculating Salary Statistics by Job Role (Top 10):")

salary_stats = analyzer.calculate_salary_stats()

print("\n--- Salary Statistics (Top 10 Roles) ---")
display(salary_stats.head(10))
```

```python
print("\n2. Counting Technology Popularity (Top 10) and Plotting:")
skill_popularity = analyzer.get_technology_popularity(top_n=10)

print("\n--- Top 10 Skill Popularity ---")
display(skill_popularity)
```

```python
print("\n3. Generating Concise Text Report (Top 5 Skills):")
report = analyzer.generate_report(top_n=5)
print(report)
```
